```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'C:/Users/ishue/OneDrive/Desktop/ML/Iris.csv'
iris_df = pd.read_csv(file_path)
iris_df
```

```python
avg_measurements = iris_df.groupby('Species').mean().drop(columns=['Id'])
colors = ['blue', 'green', 'purple', 'red']
```

```python
avg_measurements.plot(kind='bar', figsize=(12, 8), color=colors)
plt.title('Average Measurements per Species')
plt.xlabel('Species')
plt.ylabel('Measurements (cm)')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris_df, x='Species', y='SepalLengthCm', hue='Species', palette='Set3', dodge=False, legend=False)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```
